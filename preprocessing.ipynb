### Analisi prima dello split e divisione in paragrafi con split_into_paragraphs

```python
author_counts = Counter()
book_counts = {}

for author, book_id, text in load_raw_texts("RAW"):
    cleaned = clean_gutenberg(text)
    paragraphs = split_into_paragraphs(cleaned)

    n_paragraphs = len(paragraphs)
    book_counts[book_id] = n_paragraphs
    author_counts[author] += n_paragraphs

    print(f"{book_id} -> {n_paragraphs} paragrafi")

print("\nParagrafi totali per autore")
print("-"*40)
for author, count in author_counts.items():
    print(f"{author}: {count}")
```

# Split dei libri

```python
book_splits = {
    # Charlotte Bronte
    "CB-BOOK1": "train",
    "CB-BOOK2": "train",
    "CB-BOOK3": "validation",
    "CB-BOOK4": "test",

    # George Eliot
    "GE-BOOK1": "test",
    "GE-BOOK2": "train",
    "GE-BOOK3": "train",
    "GE-BOOK4": "validation",

    # Jane Austen
    "JA-BOOK1": "train",
    "JA-BOOK2": "train",
    "JA-BOOK3": "validation",
    "JA-BOOK4": "test",
}
```

```python
dataset = {
    "train": [],
    "validation": [],
    "test": []
}

for author, book_id, text in load_raw_texts("RAW"): 
    cleaned = clean_gutenberg(text) # Pulisco il testo con la funzione clean_gutenberg
    paragraphs = split_into_paragraphs(cleaned) # Ottengo i paragrafi dal testo pulito
    split = book_splits[book_id] # Ottengo lo split corrispondente al book_id

    print(f"{book_id} -> {len(paragraphs)} paragrafi")
    # conteggio paragrafi per autore

    for p in paragraphs:
        dataset[split].append({
            "text": p,
            "author": author,
            "book_id": book_id,
            "n_tokens": len(p.split())
        })
```

```python
for split_name in ["train", "validation", "test"]:
    counts = Counter(sample["author"] for sample in dataset[split_name])

    print(f"\n{split_name.upper()}")
    print("-" * 40)

    for author, count in counts.items():
        print(f"{author}: {count}")

print()
for split_name in ["train", "validation", "test"]:
    print(f"{split_name}: {len(dataset[split_name])} paragrafi totali")
```

# UNDERSAMPLING

```python
MAX_PER_AUTHOR = {"train": None, "validation": 200, "test": 200} # Non faccio undersampling sul train, ma solo su validation e test per avere un numero bilanciato di paragrafi per autore in questi split
dataset = undersample_splits(dataset, MAX_PER_AUTHOR)

print("\nDopo undersampling:")
for split_name in ["train", "validation", "test"]:
    counts = Counter(sample["author"] for sample in dataset[split_name])
    print(f"\n{split_name.upper()}")
    print("-" * 40)
    for author, count in counts.items():
        print(f"{author}: {count}")
print()
for split_name in ["train", "validation", "test"]:
    print(f"{split_name}: {len(dataset[split_name])} paragrafi totali")
```
